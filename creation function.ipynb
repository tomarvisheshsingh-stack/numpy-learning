```python
import numpy as np
```

```python
# zeros : used to create matrix with values = 0

t1 = np.zeros((5), dtype = np.int32)
print('t1 is 1D')
print(t1)
print()
t2 = np.zeros((3,3), dtype = np.int32)
print('t2 is 2D')
print(t2)
print()
t3 = np.zeros((3,3,3), dtype = np.int32)
print('t3 is 3D')
print(t3)
```

```python
t4 = np.zeros((2,2,3), dtype = np.int32)
print('t4 is 3D')
print(t4)
```

```python
# ones : used to create matrix with values = 1

t1 = np.ones((5), dtype = np.int32)
print('t1 is 1D')
print(t1)
print()
t2 = np.ones((3,3), dtype = np.int32)
print('t2 is 2D')
print(t2)
print()
t3 = np.ones((3,3,3), dtype = np.int32)
print('t3 is 3D')
print(t3)
```

```python
# eye : used to create identity matrix

t = np.eye((4), dtype=np.int32)
print(t)
```

```python
# empty : used to generate random value but they are tending towards 0

tt = np.empty(5)
print(tt)
```

```python
# arange : used to create array with specified evenly spaced values within given interval

ar = np.arange(1,11,5)
print(ar)
print()
a2 = np.arange(1,100,2)
print(a2)
```

```python
# linspace : Creates an array with a specified number of evenly spaced values over a given interval.

v, d = np.linspace(1, 10, 5, retstep = True)
print('Array :',  v)
print('diff :', d) 
```

```python
# random.rand : creates array with random values between 0 to 1

r = np.random.rand(5)
print(r)
```

```python
r2 = np.random.randint(44,48)
print(r2)
```

```python
a1 = np.array([10,3,5,6])
print(a1.ndim)
print(a1.shape)
print(a1.dtype)
print(a1.size)
```

```python
# copy(arr) : used to copy a array into another

a1 = np.array([10,3,5,6])
a2 = np.copy(a1)
print(a2)
```

```python
a1 = np.array([10,3,5,6,11])

print(a1)
a2 = np.copy(a1)
print(a2)



a1[2]=500
print(a2)
print(a1)
```

```python
# view : 
a = np.array([10,3,5,6])
print(a)
v = a1.view()
print(v)

a[1] = 312
print(v)
print(id(a))
print(id(v))
```

```python
# reshape : can create/update array of higher dimension

aa  = np.array([10,20,30,40,50,60])
print(aa)

d2 = aa.reshape(2,3)
print(d2)
print()
d3 = aa.reshape(2,1,3)
print(d3)
print()
print()
print()
d4 = aa.reshape(3,2,1)
print(d4)
```

```python
# ascending order
a = np.array([12,5,99,4,32,885,45])
print(a)
p = np.sort(a)
print(p)
```

```python
a = np.array([12,5,99,4,32,885,45])
print(a[::-1])
```

```python
# descending order
a = np.array([12,5,99,4,32,885,45])
k = np.sort(a)[::-1]
print(k)
```

```python
v = np.array([0,0,1,0,3,5,0,0])
# removing zeros
y = np.trim_zeros(v)
print(y)
# removing zeros from front
a = np.trim_zeros(v, trim='f')
print(a)
# removing zeros from back
b = np.trim_zeros(v, trim=  'b')
print(b)
```

```python
a1 = np.array([10,3,5,1,1])
a2 = np.array([10,3,5,6,45])
a = np.add(a1,a2)
print(a)
```
